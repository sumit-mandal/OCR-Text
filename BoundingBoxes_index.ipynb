```python
import pytesseract
import cv2
```

```python
image = cv2.imread("data/index_02.JPG")
base_image = image.copy()
```

```python
gray = cv2.cvtColor(image,cv2.COLOR_BGR2GRAY)
cv2.imwrite("temp/index_gray.jpg",gray)
```

```python
blur = cv2.GaussianBlur(gray,(7,7),0)
cv2.imwrite("temp/index_blur.jpg",blur)
```

### Threshold

```python
thresh = cv2.threshold(blur,0,255,cv2.THRESH_BINARY_INV + cv2.THRESH_OTSU)[1]
cv2.imwrite("temp/index_thresh.png",thresh)
```

```python
cv2.imshow("temp/index_thresh.png",thresh)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

### Dialation

```python
kernel = cv2.getStructuringElement(cv2.MORPH_RECT,(3,13))
cv2.imwrite("temp/index_kernel.png",kernel)
```

```python
cv2.imshow("temp/index_kernel.png",kernel)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
dilate = cv2.dilate(thresh,kernel,iterations=1)
cv2.imwrite("temp/index_dilate.png",dilate)
```

```python
cv2.imshow("temp/index_dilate.png",dilate)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
cnts = cv2.findContours(dilate,cv2.RETR_EXTERNAL,cv2.CHAIN_APPROX_SIMPLE)
```

```python
cnts
```

```python
cnts = cnts[0] if len(cnts) == 2 else cnts[1]
```

```python
cnts
```

```python
cnts = sorted(cnts,key = lambda x: cv2.boundingRect(x)[0])
```

```python
cnts
```

```python
for c in cnts:
    x,y,w,h = cv2.boundingRect(c)
    if h > 200 and w > 20:
        roi = image[y:y+h,x:x+h]
        cv2.imwrite("temp/index_roi.png",roi)
        cv2.rectangle(image,(x,y),(x+w , y+h), (36,255,12) ,2)
    
cv2.imwrite("temp/index_bbox.png",image)
```

```python
cv2.imshow("index_roi",roi)
# cv2.imshow("index_bbox",image)
cv2.waitKey(0)
cv2.destroyAllWindows()
```
